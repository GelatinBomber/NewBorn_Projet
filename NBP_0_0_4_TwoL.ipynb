```python
# Set seed
tf.random.set_seed(5)
```

```python
# 데이터 불러오기
wave_train = np.loadtxt('abc_wave_counting2.csv', delimiter=',', skiprows=1, dtype=np.float32)

# X
x_train = wave_train[:, :-1]
# Y
y_train = wave_train[:, [-1]]


# 행렬 확인
print(x_train.shape, y_train.shape)

# 슬라이싱 확인
print(x_train[:3])
```

```python
# class의 개수 설정
nb_classes = 1
```

```python
# 퍼셉트론(Perceptron) 개수 정하기
percep0 = x_train.shape[1]
percep1 = 64
```

```python
# Layer 1 : Hidden Layer

# Weight & Bias
W1 = tf.Variable(tf.random.normal([percep0,percep1]), name='weight1')
b1 = tf.Variable(tf.random.normal([percep1]), name='bias1')

def layer1(X):
    return  tf.nn.relu(tf.matmul(X,W1) + b1 )
```

```python
# Layer 2 : Output Layer

# Weight & Bias
W2 = tf.Variable(tf.random.normal([percep1,nb_classes]), name='weight2')
b2 = tf.Variable(tf.random.normal([nb_classes]), name='bias2')

# Linear Regression
def hypothesis(X) :
    return tf.matmul(layer1(X),W2) + b2
```

```python
# 비용 함수(cost function) : (H(x) - y)^2 의 평균
def cost_func():
    cost = tf.reduce_mean(tf.square(hypothesis(x_train) - y_train))
    return cost
```

```python
# 경사 하강법(Gradient descent)
optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01)
```

```python
# 학습 시작
print('*'*5,'Start Learnimg!!')
for step in range(8500):
    optimizer.minimize(cost_func,var_list=[W1,b1,W2,b2])
                    # 비용 함수 ,     Weight, bais 추가하기
    if step % 500 == 0:
        print(f'{step:0>4} cost :[ {cost_func().numpy():0.2f} ]')

print(f'{step:0>4} cost :[ {cost_func().numpy():0.2f} ]')
print('*'*5,'Learnimg Finished!!')
```

```python
# 경사 하강법(Gradient descent)
optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001)
```

```python
# 학습 시작
print('*'*5,'Start Learnimg!!')
for step in range(2000):
    optimizer.minimize(cost_func,var_list=[W1,b1,W2,b2])
                    # 비용 함수 ,     Weight, bais 추가하기
    if step % 500 == 0:
        print(f'{step:0>4} cost :[ {cost_func().numpy():0.2f} ]')

print(f'{step:0>4} cost :[ {cost_func().numpy():0.2f} ]')
print('*'*5,'Learnimg Finished!!')
```

```python
# 회귀 계수 출력
print('Weight =', W2.numpy())
print('Bias =', b2.numpy())
print()
```

```python
# 정확도 측정 : RMSE(Root Mean Squared Error)

def get_rmse(y_test,preds): # preds는 predicts의 약자
    squared_error = 0 # 오차를 쌓아야 하니까 변수 초기화.
    for k,_ in enumerate(y_test): # 사용하지 않을 변수는 _로 한다. k는 index를 가져온 것.
        squared_error += (preds[k] - y_test[k])**2
    mse = squared_error/len(y_test)
    rmse = np.sqrt(mse)
    return rmse[0]

# 학습한 데이터를 그대로 검증 데이터로 사용한 경우
x_test = x_train
y_test = y_train

preds = hypothesis(x_test).numpy()
print('RMSE:',get_rmse(y_test,preds))
```
