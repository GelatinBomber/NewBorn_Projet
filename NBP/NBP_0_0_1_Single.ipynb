```python
# import Library
import tensorflow as tf
import numpy as np

# Colab data upload
from google.colab import files
uploaded = files.upload()
```

```python
# 데이터 불러오기

wave_train = np.loadtxt('abc_wave_counting2.csv', delimiter=',', skiprows=1, dtype=np.float32)
                                                                    # data의 type을 처음부터 설정해서 가져옴.

# X
x_train = wave_train[:, :-1] # 마지막 컬럼을 제외
            #[행, 열]
    # print(data_train)
    # print(x_train)
        # 두 아이를 비교해서 어떻게 가져왔는지 확인.

# Y
y_train = wave_train[:, [-1]] # 마지막 컬럼만 2차원으로 추출
          # [;, -1:] 이렇게도 가능.
# 행렬 확인
print(x_train.shape, y_train.shape)

# 슬라이싱 확인
print(x_train[:1])
```

```python
# 변수 초기화 : weight, bias

tf.random.set_seed(5) # seed 고정

            # 행렬의 내적곱
                  #    X     *    W   =    Y
                  # (m, n)   *  (n, L) = (m, L)
                  # (11, 8) *  (_, _) = (11, 1)
W = tf.Variable(tf.random.normal([8,1]), name='weight')
                          # 정규분포로 만듦

b = tf.Variable(tf.random.normal([1]), name='bias')
```

```python
# 예측 함수(hypothesis) : H(x) = W*X + b

def hypothesis(X) :
    return tf.matmul(X,W) + b
```

```python
# 비용 함수(cost function) : (H(x) - y)^2 의 평균
# tf.square() : 제곱

def cost_func():
    cost = tf.reduce_mean(tf.square(hypothesis(x_train) - y_train))
    return cost
```

```python
# 경사 하강법(Gradient descent)
optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01)
```

```python
# 학습 시작
print('*'*5,'Start Learnimg!!')
for step in range(19000):
    optimizer.minimize(cost_func,var_list=[W,b])
                    # 비용 함수 ,     Weight, bais
    if step % 500 == 0:
        print(f'{step:0>4} cost :[ {cost_func().numpy():0.2f} ]')

print(f'{step:0>4} cost :[ {cost_func().numpy():0.2f} ]')
print('*'*5,'Learnimg Finished!!')
```

```python
# 정확도 측정 : RMSE(Root Mean Squared Error)

def get_rmse(y_test,preds): # preds는 predicts의 약자
    squared_error = 0 # 오차를 쌓아야 하니까 변수 초기화.
    for k,_ in enumerate(y_test): # 사용하지 않을 변수는 _로 한다. k는 index를 가져온 것.
        squared_error += (preds[k] - y_test[k])**2
    mse = squared_error/len(y_test)
    rmse = np.sqrt(mse)
    return rmse[0]

# 학습한 데이터를 그대로 검증 데이터로 사용한 경우
x_test = x_train
y_test = y_train

preds = hypothesis(x_test).numpy()
print('RMSE:',get_rmse(y_test,preds))
```

```python
# 회귀 계수 출력
print('Weight =', W.numpy())
print('Bias =', b.numpy())
```
